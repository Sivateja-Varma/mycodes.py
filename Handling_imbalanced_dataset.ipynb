#**Data is highly im-balanced**

```python
legit_transactions = df[df['Class'] == 0]
fraud_transactions = df[df['Class'] == 1]
```

```python
print(legit_transactions.shape)
fraud_transactions.shape
```

#**Undersampling**

build a sample dataset containing similar distribution of legit and fraud transactions

number of fraud transactions -> 492

```python
legit_sample = legit_transactions.sample(n=492)
```

```python
legit_sample.shape
```

```python
new_dataset = pd.concat([legit_sample,fraud_transactions],axis=0)
```
