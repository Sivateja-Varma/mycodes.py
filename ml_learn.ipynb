# **Matplotlib:**


*   Useful for making plots



```python
#importing matplotlib
import matplotlib.pyplot as plt
```

```python
#import numpy for data needed for the plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

**Plotting the data**

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
#to customize the names on the graph
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#plot a prabola
x=np.linspace(-10,10,20)
y=x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
#to customize the line to whatever u want
plt.plot(x,y,'r+')
plt.show()
```

```python
#plot more than one line in a simgle graph
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g')
plt.plot(x,np.cos(x),'r')
plt.show()
```

### Bar plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
language = ["telugu" , "hindi" , "english"]
people =[45,150,35]
plt.bar(language,people)
plt.xlabel("language")
plt.ylabel("people")
plt.title("language vs people")
plt.show()
```

# **pie plot**

```python
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
movies=['horror','action','drama']
movies_watched=[100,200,300]
plt.pie(movies_watched,labels=movies,autopct='%1.1f%%') # need to use autopct to show numbers and add % signs
plt.title('movies watched')
plt.show()
```

#scatter plot#

```python
fig3=plt.figure()
ax=fig3.add_axes([0,0,1,1])
z = np.linspace(0,10,30)
x = np.sin(z)
y = np.cos(z)
plt.scatter(z,x,color = 'g')
plt.scatter(z,y,color = 'b')
plt.show()
```
